Объединить таблицы при помощи SQL, так чтобы в таблице Role вместо profession_id и actor_id были названия профессий героев и имена актеров, которые их играют

```python
pd.read_sql(
    """
    SELECT r.name_role as name,
    p.profession as profession,
    a.name_actor as name_actor
    FROM role AS r
    JOIN profession AS p ON p.id = r.profession
    JOIN actor AS a ON a.id = r.actor
    """,
    sql)
```

Объединить таблицы при помощи SQL, так чтобы в таблице Seasons вместо season_id и role_id были названия сезонов и имена героев, которые в них присутствуют

```python
pd.read_sql(
    """
    SELECT s.id as id,
    t.season as season,
    r.name_role as hero
    FROM Seasons AS s
    JOIN TV_series AS t ON t.id = s.season_id
    JOIN role AS r ON r.id = s.role_id
    """,
    sql)
```

Объединить таблицы при помощи SQL, так чтобы в таблице Seasons вместо season_id и role_id были названия сезонов и имена героев, которые в них присутствуют и вывести только те сезоны, в которых снималась Маим Бялик

```python
pd.read_sql(
    """
    SELECT s.id as id,
    t.season as season,
    r.name_actor as actor
    FROM Seasons AS s
    JOIN TV_series AS t ON t.id = s.season_id
    JOIN actor AS r ON r.id = s.role_id
    
    WHERE actor = 'Маим Бялик'
    GROUP BY season
    """,
    sql)
```
